## Training

```python
%cd ~/work/PaddleSeg/Matting
!export CUDA_VISIBLE_DEVICES=0
!python tools/train.py \
       --config configs/ppmatting/ppmatting-hrnet_w18-human_512.yml \
       --do_eval \
       --use_vdl \
       --save_interval 50 \
       --num_workers 1 \
       --save_dir output \
       --iters 50 \
       --learning_rate 0.0001
```

### Evaluation

```python
!export CUDA_VISIBLE_DEVICES=0
!python tools/val.py \
       --config configs/ppmatting/ppmatting-hrnet_w18-human_512.yml \
       --model_path output/best_model/model.pdparams \
       --save_dir ./output/results \
       --save_results
```

### Forecast

```python
!export CUDA_VISIBLE_DEVICES=0
!python tools/predict.py \
    --config configs/ppmatting/ppmatting-hrnet_w18-human_512.yml \
    --model_path output/best_model/model.pdparams \
    --image_path data/PPM-100/val/fg/ \
    --save_dir ./output/results \
    --fg_estimate True
```

--fg_estimate False

The foreground estimation function can be turned off, which can improve the prediction speed, but the rgba image quality will be reduced



### Background Replacement

```python
!pwd
!export CUDA_VISIBLE_DEVICES=0
!python tools/bg_replace.py \
    --config configs/ppmatting/ppmatting-hrnet_w18-human_512.yml \
    --model_path output/best_model/model.pdparams \
    --image_path data/PPM-100/val/fg/13179159164_1a4ae8d085_o.jpg \
    --background g \
    --save_dir ./output/results \
    --fg_estimate True
```

--background

You can pass in the background image path, or choose one of ('r', 'g', 'b', 'w'), representing red, green, blue, white background, if not provided, use green as the background .
