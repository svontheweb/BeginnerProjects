### Forecast

modnet-mobilenetv2

```python
!mkdir logs
!export CUDA_VISIBLE_DEVICES=0
!time python3.8 tools/predict.py \
    --config configs/modnet/modnet-mobilenetv2.yml \
    --model_path modnet-mobilenetv2.pdparams \
    --image_path data/PPM-100/val/fg/ \
    --save_dir ./output/results_modnet-mobilenetv2 \
    --fg_estimate True \
 --device gpu 2>&1 | tee ./logs/modnet-mobilenetv2.log
```

--fg_estimate False

The foreground estimation function can be turned off, which can improve the prediction speed, but the rgba image quality will be reduced



modnet-resnet50

```python
!time python3.8 tools/predict.py \
    --config configs/modnet/modnet-resnet50_vd.yml \
    --model_path modnet-resnet50_vd.pdparams \
    --image_path data/PPM-100/val/fg/ \
    --save_dir ./output/results_modnet-resnet50  \
    --fg_estimate True \
 --device gcu 2>&1 | tee logs/MODNet-ResNet50_vd.log
```

modnet-hrnet_w18

```python
!time python3.8 tools/predict.py \
    --config configs/modnet/modnet-hrnet_w18.yml \
    --model_path modnet-hrnet_w18.pdparams  \
    --image_path data/PPM-100/val/fg/ \
    --save_dir ./output/results_modnet-hrnet_w18 \
    --fg_estimate True \
 --device gcu 2>&1 | tee logs/MODNet-HRNet_W18.log
```
